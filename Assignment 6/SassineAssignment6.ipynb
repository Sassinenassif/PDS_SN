What's the number of movies in 2020 with full information?

#### Exercise 5

and what has been the average in terms of releases from 2010.

```python
average_releases_2010 = df[(df['release_year'] >= 2010)].groupby('release_year')['title'].count().mean()

average_releases_2010 = round(average_releases_2010)

print(f'There was {average_releases_2010} releases on average since 2010')
```

#### Exercise 1
Calculate Gender-Based Survival Percentage

```python
path = '/Users/sassine/Desktop/PythonCourse/PDS_SN/Assignment 6/train_and_test2.csv'

df = pd.read_csv(path)
```

```python
gender_survival_percentage = df.groupby('Sex')['2urvived'].agg(['sum','count'])

gender_survival_percentage = gender_survival_percentage['sum']/gender_survival_percentage['count']



print (f'The survival percentage for Male is {round(gender_survival_percentage[0]*100)}% and for Female {round(gender_survival_percentage[1]*100)}%')

```

#### Exercise 2
Calculate Survival Percentage Grouped by Gender and Class

```python
class_gender_survival = df.groupby(['Pclass', 'Sex'])['2urvived'].agg(['sum', 'count'])
class_gender_survival['Survival Percentage'] = (class_gender_survival['sum'] / class_gender_survival['count']) * 100
class_gender_survival.drop(columns= ['sum','count'])
```
