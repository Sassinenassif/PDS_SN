### Session 8: Plotting


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Student Name': ['Amelie', 'Edgar', 'Jordi', 'Mar√ßal', 'Pep', 'Jaume', 'Marco', 
                     'Ludmila', 'Bastian', 'Marc', 'Gerardo', 'Javier', 'Frank', 'Julia'],
    'Course': ['Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science'],
    'Grade': [85, 92, 78, 88, 91, 84, 73, 95, 90, 80, 85, 87, 89, 92],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Female'],
    'Study Time': [15, 20, 10, 8, 18, 14, 11, 16, 19, 13, 16, 12, 17, 15],
    'Age': [22, 21, 23, 22, 20, 24, 22, 21, 23, 21, 22, 22, 21, 20]
}

df = pd.DataFrame(data)
```

### Exercise 1
Create a lineplot showing how Study Time varies by Student Name. Which student has the highest study time?

```python
plt.plot(df["Student Name"], df["Study Time"], marker = "o", color = 'pink' )
plt.title('Student Study Time', fontsize=14)
plt.xlabel("Student Name")
plt.ylabel("Study Time (h)")
plt.xticks(rotation = 45, fontsize = 8) #rotate student names so they don't overlap

plt.show()
```

```python
max_study_time = df[df['Study Time'] == df['Study Time'].max()]['Student Name'].iloc[0]
print(f"The student with the highest study time is: {max_study_time}")
```

### Exercise 2
Plot a histogram (histplot) of Grade and determine which grade range has the highest frequency of students.

```python
sns.histplot(df['Grade'], bins=10, kde=False)
plt.title("Histogram of Grades")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
grade_range = pd.cut(df['Grade'], bins=10).value_counts().idxmax()
print(f"The grade range with the highest frequency is: {grade_range}")
```

### Exercise 3

Create an ECDF plot (*ecdfplot*) for **Grade**. What is the percentage of students scoring less than 85?

```python


sns.ecdfplot(data=df, x="Grade", color="black", linewidth=2)

plt.title("Grade")

plt.show()
```

```python
percentage_less_than_85 = (df['Grade'] < 85).mean() * 100
print(f"The percentage of students scoring less than 85 is: {percentage_less_than_85:.2f}%")
```

### Exercise 4
Create a *stripplot* showing **Grade** distribution for each **Course**. Which course has the most spread in grades?


```python
sns.stripplot(data=df, x="Grade", y="Course", hue="Course")
plt.title("Grade Distribution by Course")
plt.grid()

plt.show()
```

```python
spread = df.groupby('Course')['Grade'].std()
course_with_max_spread = spread.idxmax()
print(f"The course with the most spread in grades is: {course_with_max_spread}")
```

### Exercise 5
Create a *swarmplot* to show the relationship between Gender and **Study Time**. Which gender has a higher average study time?

```python
sns.swarmplot(data=df, x="Study Time", y="Gender", hue="Gender")
plt.title("Study Time Distribution by Gender")

plt.show()
```

```python
avg_studytime_gen = df.groupby('Gender')['Study Time'].mean()
highest_studytime_gen = avg_studytime_gen.idxmax()
print(f"{highest_studytime_gen} has the highest average study time")
```

### Exercise 6
Plot a *pointplot* to show the average **Grade** for each Course. Which course has the highest average grade?

```python
sns.pointplot(x='Course', y='Grade', data=df)
plt.title("Average Grade by Course")
plt.xlabel("Course Name")  
plt.ylabel("Average Grade") 

# Rotate x-axis labels for better readability (Chatgpt assisted to make it more readable)
plt.xticks(rotation=45)


plt.show()




```

```python
avg_grade_by_course = df.groupby('Course')['Grade'].mean()
course_with_highest_avg_grade = avg_grade_by_course.idxmax()
print(f' {course_with_highest_avg_grade} has the highest average grade')
```
